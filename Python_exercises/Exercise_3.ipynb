# Exercise 1: 

Write a Python program to implement the following class hierarchy<br>


```python
# Creating Asset Class
class Asset:
    def __init__(self,descriptor,date,currentValue):
        self.descriptor = descriptor
        self.date = date
        self.currentValue = currentValue
        
# Creating the three derived classes from Aseet class

class Stock(Asset):
    def __init__(self,descriptor,date,currentValue,numShares,sharePrice,asset):
        super().__init__(descriptor,date,currentValue)
        self.numShares = numShares
        self.sharePrice = sharePrice
        self.asset = asset
        
class Bond(Asset):
    def __init__(self,descriptor,date,currentValue,interestRate,asset):
        super().__init__(descriptor,date,currentValue)
        self.interestRate = interestRate
        self.asset = asset
        
class Savings(Asset):
    def __init__(self,descriptor,date,currentValue,interestRate,asset):
        super().__init__(descriptor,date,currentValue)
        self.interestRate = interestRate
        self.asset = asset
```

# Exercise 2:

Create a class Polygon which has an attribute `number_of_sides`. In the class `Polygon` define four functions: 
    
- `whoAmI(self)` --> function that answers according to the number of sides which Polygon it is (3 = Triangle, 4 = Rectangle, else Polygon)


- `howManySides(self)` --> A getter for the attribute number_of_sides


- `area` --> returns `"No area"` if the function is in the class `Polygon`. For the other shapes it returns the area of the specific polygon


- `perimeter` --> returns `"No perimeter"` if the function is in the class Polygon. For the other shapes it returns the perimeter of the specific polygon
    
Create two more classes `Triangle` and `Rectangle` that inherit the methods from the class `Polygon`. Each of the subclasses should have their own constructor and construct attributes respective of the number of sizes for the polygon.

Create `area` and `perimeter` methods in each of the subclasses to override (or overload) the `area` and `perimeter` methods in the super class

Note: 
- Area of Triangle with a, b and c: https://www.cuemath.com/measurement/area-of-triangle/
- Perimeter of Triangle with a, b and c: a + b + c

```python
import math as m

class Polygon:
    def __init__(self,number_of_sides):
        self.number_of_sides = number_of_sides
    
    # Function that returns what type of Polygon it is
    def whoAmI(self):
        if self.number_of_sides == 3:
            return 'Triangle'
        elif self.number_of_sides == 4:
            return 'Rectangle'
        elif self.number_of_sides > 4:
            return 'Polygon'
        else:
            return 'That is not a polygon'
        
    # Getter that returns how many sides the Polygon has
    def howManySides(self):
        return self.number_of_sides
    
    # These two fucntions will be overried in the Other child clasees
    def area(self):
        return 'No Area'
    
    def perimeter(self):
        return 'No Perimeter'

# Creating child class Triangle with parent class Polygon
class Triangle(Polygon):
    # Creting a constructor
    def __init__(self,a,b,c):
        super().__init__(3)
        self.a = a
        self.b = b
        self.c = c
        
    # Overriding area method suitable for class Triangle
    def area(self):
        s = (self.a + self.b + self.c)/2
        return m.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))
    
    # Overriding perimeter method suitable for class Triangle
    def perimeter(self):
        return self.a + self.b + self.c
    
# Creating child class Rectangle with parent class Polygon
class Rectangle(Polygon):
    def __init__(self, l, w):
        super().__init__(4)
        self.l = l
        self.w = w
        
    # Overriding area method suitable for class Rectangle
    def area(self):
        return self.l * self.w
    
    # Overriding perimeter method suitable for class Rectangle
    def perimeter(self):
        return 2 * (self.l + self.w)
    
# Creating instances for the class Polygon

p = Polygon(2)

print(p.whoAmI())
print(p.howManySides())   

   

# Creting instance for the class Triangle
t = Triangle(5,6,10)

# Checking if the lenght of the three sides make a triangle
if (t.a + t.b > t.c and t.a + t.c > t.b and t.b + t.c > t.a):
    
    print(t.whoAmI())
    print(f'I have {t.howManySides()} sides')
    
    print(f'The area of the Triangle is: {t.area()} with its perimeter: {t.perimeter()}')
          
else:
    print("This is not a triangle")


# Creating instance for the class Recatangle
r = Rectangle(2,4)

print(r.whoAmI())
print(f'I have {r.howManySides()} sides')

print(f'The area of the Rectangle is: {r.area()} with its paremeter: {r.perimeter()}')
```

```python
# Creating a base class Vehicle
class Vehicle:
    def __init__(self,CC):
        self.color = CC
        
# Creating the 4 derived classes from Vehicle (Car,Bike,Bus,Truck)       
class Car(Vehicle):
    def __init__(self,CC,NW,HP):
        super().__init__(CC)
        self.number_of_wheels = NW
        self.horse_power = HP
        
        
class Bike(Vehicle):
    def __init__(self,CC,NW,HP):
        super().__init__(CC)
        self.number_of_wheels = NW
        self.horse_power = HP
        
class Bus(Vehicle):
    def __init__(self,CC,NW,HP):
        super().__init__(CC)
        self.number_of_wheels = NW
        self.horse_power = HP  
        
        
class Truck(Vehicle):
    def __init__(self,CC,NW,HP):
        super().__init__(CC)
        self.number_of_wheels = NW
        self.horse_power = HP
        
        
# Creating the derived clasess from Car class (ElectricCar,CombustionEngineCar)

class ElectricCar(Car):
    def __init(self, CC, NW, HP):
        super().__init__(CC,NW,HP)
        self.co_2_emmisions = 0
        
        
class CombustionEngineCar(Car):
    def __init__(self,CC, NW, HP,co2e):
        super().__init__(CC,NW,HP)
        self.co_2_emmisions = co2e
        
        
# Creating intances for the classes:

# Creating instance for the class Vehicle
v1 = Vehicle('purple')
v2 = Vehicle('blue')
print(v1.color)
print(v2.color)

# Creating instances for the clas Car for the other thee classes that derrive from Vehicle the instances are made the same way
c1 = Car(v1.color,4,1111)
c2 = Car(v2.color,4,2222)
print(f'The car with the color {c1.color} has {c1.number_of_wheels} wheels and {c1.horse_power} horse power')
print(f'The car with the color {c2.color} has {c2.number_of_wheels} wheels and {c2.horse_power} horse power')

# Creating instances for the class ElectricCar
ec = ElectricCar(c1.color,c1.number_of_wheels,c1.horse_power)
print(f'The Electric car is {ec.color} with {ec.number_of_wheels} wheels and {ec.horse_power} horse power')

# Creating instances for the class CombustionEngineCar
cec = CombustionEngineCar(c2.color,c2.number_of_wheels,c2.horse_power,12)
print(f'The Combustion engine car is {cec.color} with {cec.number_of_wheels} wheels and {cec.horse_power} horse power with {cec.co_2_emmisions} emmisions')
```

#### Using `property()` function

```python
# Creating a base class called Person
class Person:
    def __init__(self,fname,lname,age):
        self.__fname = fname
        self.__lname = lname
        self.__age = age
        
    # Set method setFirstName
    def __setFirstName(self,fname):
        self.__fname = fname
        
    # Set method setLastName
    def __setLastName(self,lname):
        self.__lname = lname
        
    # Set method setAge
    def __setAge(self,age):
        self.__age = age
        
    # Get method getFirstName
    def __getFirstName(self):
        return self.__fname
    
    # Get mehtod getLastName
    def __getLastName(self):
        return self.__lname
    
    # Get method getAge
    def __getAge(self):
        return self.__age
    
    firstName = property(__getFirstName,__setFirstName)
    lastName = property(__getLastName,__setLastName)
    age = property(__getAge,__setAge)
        
class Student(Person):
    def __init__(self,fname,lname,age,student_id):
        super().__init__(fname,lname,age)
        self.__student_id = student_id
    
    # Set method setStId
    def __setStId(self,st_id):
        self.__student_id = st_id
        
    # Get method getStId
    def __getStId(self):
        return self.__student_id
    
    student_id = property(__getStId,__setStId)
```

```python
# Making nistances for the two classes
p1 = Person('Teodora','Iloska',23)
print(f'My name is {p1.firstName} {p1.lastName}, and I have {p1.age} years.')
p1.age = 25
print(f'My name is {p1.firstName} {p1.lastName}, and I have {p1.age} years.')

s1 = Student('John','Doe',19,'11111')
print(f'His name is {s1.firstName} {s1.lastName}, he is {s1.age} years old. His index number is {s1.student_id}.')
s1.student_id = '22222'
s1.firstName = 'Peter'
print(f'His name is {s1.firstName} {s1.lastName}, he is {s1.age} years old. His index number is {s1.student_id}.')
```

#### Using decorators

```python
# Creating a base class called Person
class Person:
    def __init__(self,fname,lname,age):
        self.__fname = fname
        self.__lname = lname
        self.__age = age
    
    @property
    def firstName(self):
        return self.__fname
    
    @firstName.setter
    def firstName(self,fname):
        self.__fname = fname
    
    @property
    def lastName(self):
        return self.__lname
    
    @lastName.setter
    def lastName(self,lname):
        self.__lname = lname
    
    @property
    def age(self):
        return self.__age
    
    @age.setter
    def age(self,age):
        self.__age = age
        
        
class Student(Person):
    def __init__(self,fname,lname,age,student_id):
        super().__init__(fname,lname,age)
        self.__student_id = student_id
        
    @property
    def student_id(self):
        return self.__student_id
    
    @student_id.setter
    def student_id(self,st_id):
        self.__student_id = st_id
        
```

```python
# Making nistances for the two classes
p1 = Person('Teodora','Iloska',23)
print(f'My name is {p1.firstName} {p1.lastName}, and I have {p1.age} years.')
p1.age = 25
print(f'My name is {p1.firstName} {p1.lastName}, and I have {p1.age} years.')

s1 = Student('John','Doe',19,'11111')
print(f'His name is {s1.firstName} {s1.lastName}, he is {s1.age} years old. His index number is {s1.student_id}.')
s1.student_id = '22222'
s1.firstName = 'Peter'
print(f'His name is {s1.firstName} {s1.lastName}, he is {s1.age} years old. His index number is {s1.student_id}.')
```

```python
class Employee:
    def __init__(self,emp_id,fname,lname,salary):
        self.__emp_id = emp_id
        self.__fname = fname
        self.__lname = lname
        self.__salary = salary
        
    def getID(self):
        return self.__emp_id
    
    def getFirstName(self):
        return self.__fname
    
    def getLastName(self):
        return self.__lname
    
    def getName(self):
        return f'{self.__fname} {self.__lname}'
    
    def getSalary(self):
        return self.__salary
    
    def setSalary(self,salary):
        self.__salary = salary
        
    def getAnnualSalary(self):
        return self.__salary * 12
    
    def raiseSalary(self,per):
        self.__salary += (self.__salary * per/100)
        return self.__salary
    
    def toString(self):
        return f'Employee [id = {self.__emp_id}, name = {self.getName()}, salary = {self.__salary} ]'
```
