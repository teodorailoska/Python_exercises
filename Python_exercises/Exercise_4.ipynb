### Exercise 1

Write a map function that adds plus 5 to each item in the list `[1,5,7,2,4,6,7]`.

```python
l = [1,5,7,2,4,6,7]

def plus_5(elem):
    return elem + 5

fin_list = list(map(plus_5,l))
fin_lambda = list(map(lambda x: x+5,l))

print(fin_lambda)
print(fin_list)
```

### Exercise 3

Using `filter` and `lambda` count the even and odd numbers in the following list `[1,5,7,2,4,6,7]`

```python
lst = [1,5,7,2,4,6,7]

even = len(list(filter(lambda x: x % 2 == 0,lst)))
odd = len(list(filter(lambda x: x % 2 != 0,lst)))

print(even)
print(odd)
```

### Exercise 5

Given two arrays `x_arr = [1,2,3,4,5,6,7]` and `y_arr = [2,7,9,1]`
write a **reduce + filter** combination so that you add up the numbers that are present in both arrays

```python
from functools import reduce
# Za doma
x_arr = [1,2,3,4,5,6,7]
y_arr = [2,7,9,1]

print(x_arr[-1])
print(y_arr)
# Prvo gi filtritram listite za tie elemnti sto gi ima i vo x_arr i vo y_arr
filt = list(filter(lambda x : x in y_arr,x_arr))

# Potoa so reduce go naogjam nivniot zbir
print(reduce(lambda x,y: x+y,filt))
```

### Exercise 6

Using filter and reduce return the product of the elements from the first list that are not present in the second list.  
`l1 = [5, 3, 3, 9, 10, 2, 7]`

`l2 = [0, 1, 3, 7, 10, 1, 7]` 

```python
from functools import reduce
# Za doma
l1 = [5, 3, 3, 9, 10, 2, 7]
l2 = [0, 1, 3, 7, 10, 1, 7]

filt = list(filter(lambda x : x not in l2,l1))
print(filt)
print(reduce(lambda x,y: x*y,filt))
```

Create two pandas Series from the following lists `[2, 4, 6, 8, 10]`, `[1, 3, 5, 7, 9]`

### Exercise 3

Create two pandas Series from the following lists `[1, 2, 3, 4, 5]` and `[4, 5, 6, 7, 8]`
Read about the `isin()` function. Use it to get the items of series A not present in series B.  
**Hint:** also use the `~` for negative reverse

```python
# Za doma
import pandas as pd

l1 = [1, 2, 3, 4, 5]
l2 = [4, 5, 6, 7, 8]
# Creatinga two panda series with the following lists
a = pd.Series(l1)
b = pd.Series(l2)

print('Series a:')
print(a)

print('Series b:')
print(b)

# I am creating a new series c that consists only the items that belong in a but don't belong in b
c = a[~a.isin(b)]
print(f'Items that are not in series b:')
print(c)
```

### Bonus Pandas Exercise:

Load the `amazon.csv` dataset. Find the following:

- Find the total amount of fires that occurred in the month of January across all years.
- Find the average number of fires per year across all years.
- Find the year in which there were the most number of fires in August.
- Find the state that had the minimum number of total fires across all years.
- Find the mean, median and standard deviation of the number of fires per year.
- Find the mean, median and standard deviation of the number of fires per state.
- Find the total amount of fires for the month of April for the state of Sao Paulo between the years of 2004 and 2014.

```python
# Ovaa bonus vezba e slicna na proektot sto ke bide naredna nedela

import pandas as pd

df = pd.read_csv('amazon.csv', sep = ';')
df
```

```python
df.info()
```

```python
# Find the total amount of fires that occurred in the month of January across all years.
first_jan = df[df["month"] == "January"]
first_jan["number_of_fires"].sum()
```

```python
df[df["month"] == "January"]["number_of_fires"].sum()
```

```python
# Find the average number of fires per year across all years.

# SQL
#select avg(number_of_fires)
#from df
#group by year

df.groupby('year')['number_of_fires'].mean()
```

```python
# Find the year in which there were the most number of fires in August.
fires_per_year_August = df[df['month'] == 'August'].groupby('year')['number_of_fires'].sum()
max_year = fires_per_year_August.idxmax()
print(fires_per_year_August)
print("The year with the most number of fires in August is",max_year)
```

```python
# Solution Valentin
df[df['month'] == 'August'].groupby('year')['number_of_fires'].sum().sort_values(ascending = False).index[0]
```

```python
# Find the state that had the minimum number of total fires across all years.
print(df.groupby('state')['number_of_fires'].sum())
df.groupby('state')['number_of_fires'].sum().idxmin()
```

```python
df.groupby('state')['number_of_fires'].sum().sort_values().index[0]
```

```python
#Find the mean, median and standard deviation of the number of fires per year
print(df.groupby('year')['number_of_fires'].mean())
print(df.groupby('year')['number_of_fires'].median())
print(df.groupby('year')['number_of_fires'].std())
```

```python
# Find the mean, median and standard deviation of the number of fires per state.
print(df.groupby('state')['number_of_fires'].mean())
print(df.groupby('state')['number_of_fires'].median())
print(df.groupby('state')['number_of_fires'].std())
```

```python
# Find the total amount of fires for the month of April for the state of Sao Paulo between the years of 2004 and 2014.

total_amount = df[(df['month'] == 'April') & 
                  (df['state'] == 'Sao Paulo') & 
                  (df['year'] >= 2004) & 
                  (df['year'] <= 2014)]['number_of_fires']
print(total_amount)
print(total_amount.sum())
```

```python
# So between funkcija
total_amount = df[(df['month'] == 'April') & 
                  (df['state'] == 'Sao Paulo') & 
                  (df['year'].between(2004,2014))]['number_of_fires']
print(total_amount)
print(total_amount.sum())
```
